```python
data_path = '/explore/nobackup/projects/above/misc/ABoVE_Shrubs/development/landcover/v1.1'
```

```python
import os
from pathlib import Path
import rioxarray as rxr
import numpy as np
import geopandas as gpd
import rioxarray as rxr
import xarray as xr
```

```python
for raster_filename in rasters_list:

    landcover_filename = os.path.join(data_path, f'{Path(raster_filename).stem}.landcover.tif')

    if not os.path.exists(landcover_filename):
        continue

    print(raster_filename)
    print(landcover_filename)

    da = rxr.open_rasterio(landcover_filename, mask_and_scale=True)  # DataArray: dims=("band","y","x")
    # mask_and_scale=True applies scale/offset if present, and masks nodata
    
    # Reproject GDF to raster CRS if needed
    if gdf.crs is None:
        raise ValueError("GeoDataFrame has no CRS. Set gdf.crs before sampling.")
    if gdf.crs.to_string() != da.rio.crs.to_string():
        gdf = gdf.to_crs(da.rio.crs)
    
    # Vectorized nearest-neighbor point sampling
    xs = xr.DataArray(gdf.geometry.x.values, dims="z")
    ys = xr.DataArray(gdf.geometry.y.values, dims="z")
    
    # .sel with method="nearest" uses the raster geotransform to pick the nearest pixel
    vals = da.sel(x=xs, y=ys, method="nearest")  # dims: band, z
    
    # Attach to GDF
    if "band" in vals.dims and vals.sizes["band"] > 1:
        for i in range(vals.sizes["band"]):
            gdf[f"band_{i+1}"] = vals.isel(band=i).values
    else:
        gdf["rast_val"] = vals.values if "band" not in vals.dims else vals.isel(band=0).values
    
    # Save
    gdf.to_file("points_with_samples_v1.1.gpkg", driver="GPKG")
    gdf.drop(columns="geometry").to_csv("points_with_samples_v1.1.csv", index=False)
```

```python
from pathlib import Path
import os
import numpy as np
import geopandas as gpd
import rioxarray as rxr
import xarray as xr

# gdf already loaded
# gdf: GeoDataFrame with points and a valid gdf.crs

def sample_raster_into_gdf(gdf, raster_path, col_prefix):
    da = rxr.open_rasterio(raster_path, mask_and_scale=True)  # dims: ("band","y","x")
    if da.rio.crs is None:
        raise ValueError(f"Raster has no CRS: {raster_path}")

    # Reproject points only if needed (cheap; but we’ll also cache xs/ys per CRS)
    if gdf.crs is None:
        raise ValueError("GeoDataFrame has no CRS. Set gdf.crs before sampling.")
    if gdf.crs.to_string() != da.rio.crs.to_string():
        gdf_proj = gdf.to_crs(da.rio.crs)
    else:
        gdf_proj = gdf

    # Build x/y arrays (vectorized sampling)
    xs = xr.DataArray(gdf_proj.geometry.x.values, dims="z")
    ys = xr.DataArray(gdf_proj.geometry.y.values, dims="z")

    # Nearest neighbor (correct for categorical landcover)
    vals = da.sel(x=xs, y=ys, method="nearest")  # dims: band, z  (or just z if single-band)
    # Attach columns
    if "band" in vals.dims and vals.sizes["band"] > 1:
        for i in range(vals.sizes["band"]):
            gdf[f"{col_prefix}_b{i+1}"] = vals.isel(band=i).values
    else:
        # squeeze to 1D
        gdf[f"{col_prefix}"] = vals.values if "band" not in vals.dims else vals.isel(band=0).values

    return gdf

# --- main loop ---
for raster_filename in rasters_list:
    landcover_filename = os.path.join(
        data_path, f"{Path(raster_filename).stem}.landcover.tif"
    )
    if not os.path.exists(landcover_filename):
        continue

    print("Sampling:", raster_filename)
    print("Landcover:", landcover_filename)

    stem = Path(landcover_filename).stem  # e.g., "117408_028.landcover"
    # Make a clean, short column prefix (e.g., "lc_117408_028")
    col_prefix = "lc_" + stem.replace(".landcover", "")

    gdf = sample_raster_into_gdf(gdf, landcover_filename, col_prefix)

# Save once at the end
gdf.to_file("points_with_samples_v1.1.gpkg", driver="GPKG")
gdf.drop(columns="geometry").to_csv("points_with_samples.csv", index=False)
print("Saved samples to points_with_samples.gpkg / .csv")
```

```python
import re

# Make a dict mapping col → year
col_year = {}
for col in gdf.columns:
    if col.startswith("lc_"):
        m = re.search(r"(\d{8})", col)  # find YYYYMMDD
        if m:
            year = int(m.group(1)[:4])
            col_year[col] = year

col_year
```

```python
# Suppose your reference year is in "meco2_date"
val_options["obs_year"] = gdf["meco2_date"].dt.year

# For each year, attach raster values from the same-year columns
for year in sorted(set(col_year.values())):
    # All raster cols for this year
    cols = [c for c,y in col_year.items() if y == year]
    if not cols:
        continue
    
    mask = val_options["obs_year"] == year
    if not mask.any():
        continue
    
    # Merge same-year rasters: take first non-null among them
    val_options.loc[mask, f"lc_val_{year}"] = (
        val_options.loc[mask, cols].bfill(axis=1).iloc[:, 0]
    )
val_options
```

```python
# For each row, select the raster value corresponding to its obs_year
val_options["lc_val_match"] = val_options.apply(
    lambda row: row.get(f"lc_val_{row.obs_year}", np.nan),
    axis=1
)

val_options = val_options.dropna(subset=["lc_val_match"]).copy()

print(f"Original rows: {len(val_options)}")
print(f"Remaining after drop: {len(val_options)}")

val_options
```

```python
final_map = {
    "moist tundra": 0,
    "low shrub": 1,
    "tall shrub": 2,
    "barren": 3,
    "water": 4,        # both lake & river/ocean → 4
    "spruce": 5,
    "other": 6,        # cloud shadow + snow + misc collapsed
    "wet meadow": 7,   # includes marsh
    "marsh": 7,
}

old_to_final = {
    0: 0,   # moist tundra
    1: 1,   # low shrub
    2: 2,   # tall shrub
    3: 3,   # barren
    4: 4,   # water (lake)
    5: 4,   # river/ocean → 4
    6: 5,   # spruce
    7: 6,   # cloud shadow
    8: 6,   # snow → 6
    9: 7,   # marsh/wet meadow
}

val_options["landcover_reference"] = (
    val_options["meco2_name"].str.lower().str.strip().map(final_map)
)

val_options["landcover_prediction"] = (
    val_options["lc_val_match"].round().astype(int).replace(old_to_final)
)

val_options
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with NaNs in either column (safety check)
df = val_options[val_options['obs_year'] == 2020].dropna(subset=["landcover_reference", "landcover_prediction"]).copy()

y_true = df["landcover_reference"].astype(int).values   # from your observations
y_pred = df["landcover_prediction"].astype(int).values  # from raster predictions

# --- Metrics ---
acc = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="weighted", zero_division=0)
rec = recall_score(y_true, y_pred, average="weighted", zero_division=0)
f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
print("F1 Score :", f1)

# Detailed per-class report
print("\nClassification report:")
print(classification_report(y_true, y_pred, zero_division=0))

# --- Confusion Matrix ---
cm = confusion_matrix(y_true, y_pred, labels=np.arange(8))  # since final codes are 0–7

# Pretty plot
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=[f"{i}" for i in range(8)],
            yticklabels=[f"{i}" for i in range(8)])
plt.xlabel("Predicted")
plt.ylabel("Observed")
plt.title("Confusion Matrix (landcover)")
plt.show()
```
